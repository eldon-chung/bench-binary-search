```python
import argparse
import json
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp

def get_json(file_path):
    with open(file_path, "r") as f:
        return json.load(f)

def get_algo_name(cmd):
    return cmd.split(" ")[0].removeprefix("./build/src/algos/")

def get_summary_point(iter_idx, query_type, file_power):
    file_name = F"n-chunked-random-s-{file_power:02}-w-uint32-result.json"
    file_path = F"results/{query_type}{iter_idx}/{file_name}"

    bench_json = get_json(file_path)
    
    collection = {}
    algo_names = list_algo_names()
    for algo in algo_names:
        collection[algo] = {}
    
    for result in bench_json["results"]:
        collection[get_algo_name(result["command"])][file_power] \
            = { "mean": float(result["mean"]), 
                "stddev": float(result["stddev"]),
                "max": float(result["max"]) } 
    return collection

def get_timings_point(iter_idx, query_type, file_power):
    file_name = F"n-chunked-random-s-{file_power:02}-w-uint32-result.json"
    file_path = F"results/{query_type}{iter_idx}/{file_name}"

    bench_json = get_json(file_path)

    collection = {}
    algo_names = list_algo_names()
    for algo in algo_names:
        collection[algo] = {}
    
    for result in bench_json["results"]:
        collection[get_algo_name(result["command"])] = result["times"]
    return collection

def get_summaries_size_axis(iter_idx, query_type):
    summaries = {}
    algo_names = list_algo_names()
    for algo in algo_names:
        summaries[algo] = {}

    for file_power in range(18):
        summary_point = get_summary_point(iter_idx, query_type, file_power)
        for algo in algo_names:
            summaries[algo][file_power] = summary_point[algo][file_power]

    return summaries

def get_summaries_position_axis(file_power, query_type):
    summaries = {}
    algo_names = list_algo_names()
    for algo in algo_names:
        summaries[algo] = {}
    
    end_val = 10
    if query_type == "n":
        end_val = 20
    for iter_idx in range(end_val):
        summary_point = get_summary_point(iter_idx, query_type, file_power)
        for algo in algo_names:
            summaries[algo][iter_idx] = summary_point[algo][file_power]

    return summaries

# def get_timings_size_axis(iter_idx, query_type):
#     timings = {}

#     for file_power in range(18):
#         get_timings_point(timings, iter_idx, query_type, file_power)
#     return timings

def into_mean_array(points_dict):
    # assumes we start from 0, which we do
    arr = [0] * len(points_dict.keys()) 

    for key in points_dict.keys():
        arr[int(key)] = points_dict[key]["mean"]
    
    return arr

def list_algo_names():
    file_name = F"n-chunked-random-s-00-w-uint32-result.json"
    file_path = F"results/p0/{file_name}"

    bench_json = get_json(file_path)

    names = []
    for result in bench_json["results"]:
        names.append(get_algo_name(result["command"]))
    return names

def welch_test(algo1, algo2, query_type, file_size, position_hint):
    all_timings = get_timings_point(position_hint, query_type, file_size)
    algo1_timings = np.array(all_timings[algo1])
    algo2_timings = np.array(all_timings[algo2])
    print(F"mean 1: {np.mean(algo1_timings)} ; mean 2: {np.mean(algo2_timings)}")
    print(F"variance 1: {np.var(algo1_timings)} ; variance 2: {np.var(algo2_timings)}")
    print(F"welch t test results: ", sp.stats.ttest_ind(algo1_timings, algo2_timings, equal_var = False))
    
def plot_all_timings_size_axis(query_type, position_hint):
    algos = list_algo_names()
    plot_all_timings_size_axis_sublist(query_type, position_hint, algos)

def plot_all_timings_position_axis(query_type, file_size):
    algos = list_algo_names()
    plot_all_timings_position_axis_sublist(query_type, file_size, algos)

def plot_all_timings_size_axis_sublist(query_type, position_hint, algo_sublist):
    algos = list_algo_names()
    summaries = get_summaries_size_axis(position_hint, query_type)
    
    algo_to_mean_runtime = {}
    for algo in algos:
        algo_to_mean_runtime[algo] = []
        for file_size in range(len(summaries[algo])):
            algo_to_mean_runtime[algo].append(summaries[algo][file_size]["mean"])
    
    for algo in algos:
        if algo not in algo_sublist:
            continue
        plt.plot(np.array(algo_to_mean_runtime[algo]), label=algo)
    plt.legend()
    plt.xticks(np.arange(0, 18, 1))
    plt.xlabel("file size = 2^x * 2048 bytes")
    plt.ylabel("mean wall clock time in ms")
    plt.show()

def plot_all_timings_position_axis_sublist(query_type, file_size, algo_sublist):
    algos = list_algo_names()
    summaries = get_summaries_position_axis(file_size, query_type)
    
    algo_to_mean_runtime = {}
    for algo in algos:
        algo_to_mean_runtime[algo] = []
        for it_val in range(len(summaries[algo])):
            algo_to_mean_runtime[algo].append(summaries[algo][it_val]["mean"])
            
            
    
    max_x = 10
    for algo in algos:
        if algo not in algo_sublist:
            continue
        max_x = max(max_x, len(algo_to_mean_runtime[algo]) + 1)
        plt.plot(np.array(algo_to_mean_runtime[algo]), label=algo)
    
    plt.legend()
    plt.xticks(np.arange(1, max_x, 1))
    plt.xlabel(F"position hint (query type {query_type}; file size 2048 * 2^{file_size})")
    plt.ylabel("mean wall clock time in ms")
    ax = plt.subplot(111)
    ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
    plt.show()
```
